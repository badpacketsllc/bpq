```python
# Import the API client and libraries needed to set constants
from bpq.client import BadPacketsHttpApiClient
from datetime import datetime
from datetime import timedelta
```

```python
# Pull data from the last hour and a half
CURRENT_TIME = datetime.utcnow()
DELTA_TIME = CURRENT_TIME - timedelta(minutes=90)
DELTA_TIME = DELTA_TIME.isoformat()
```

```python
bad_packets_http_api_client = BadPacketsHttpApiClient()
```

```python
bad_packets_http_api_client.query_data(last_seen_after=DELTA_TIME)
```
